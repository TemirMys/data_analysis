## Предобработка данных

```python
df['balcony'] = df['balcony'].fillna(0)
df['balcony'].isna().sum()
```

```python
df.loc[df['parks_around3000'] == 0, 'parks_nearest'] = 0 
df.loc[df['parks_nearest'] == 0, 'parks_around3000'] = 0 


df.loc[df['ponds_around3000'] == 0, 'ponds_nearest'] = 0 
df.loc[df['ponds_nearest'] == 0, 'ponds_around3000'] = 0 

df.isna().sum()
```

```python
df = df.dropna(subset = ['floors_total', 'locality_name'])
df.isna().sum()
```

<b> По рекомендации из инструкций к проекту, я заполнил пустые значения в столбце балконов на 0, так как пропуски в данной колонке, скорее всего, означали, что балконов в квартире нет.
    <br/> Я заполнил нулями пропуски в колонках parks и ponds, где один из критериев (nearest, around3000) был равен 0. По моей логике, если человек указал значение "0" по одному из этих критериев, то, скорее всего, это должно означать равное и по второму. В итоге количество пропусков в этих четырех колонка уравнялось, и теперь их количество 5518.
    <br/> Количество пропусков в колонке floors_total составляло 86, locality_name - 48. Это даже меньше 1% от общего количества значений. Поэтому я решил дропнуть пропуски.
 

```python
df.loc[df['studio'] == True, 'kitchen_area'] = 0 
df.loc[df['studio'] == True, 'rooms'] = 0
# значение площади кухни в квартирах-студиях заменил на нулевое значение
df.loc[df['open_plan'] == True, 'kitchen_area'] = 0 
df.loc[df['open_plan'] == True, 'rooms'] = 0 
# то же самое проделал для квартир с открытой планировкой. 
```

<b> Открытая планировка: Обычно "открытыми" делают общие комнаты – кухня плавно перетекает в столовую, затем в гостиную. 
<br/> Квартира-студия: Студия — жилая недвижимость в стиле open space, то есть с открытым пространством. В ней нет межкомнатных стен и отдельных комнат. </b>

```python
median_kitchen = (df['kitchen_area']/df['total_area']).median()
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area']*median_kitchen)
df.isna().sum()
```

<b> Пустые значения в колонке площади кухни заполнил по следующей схеме: разделил площадь кухни на общую площадь. Нашел медианное значение доли кухни от общей площади. Затем заполнил пропуски произведением доли кухни на общую площадь в соответствующей строке

```python
median_living_grouped = df.groupby('rooms')['living_area'].median()
median_living = pd.DataFrame({'median_living':median_living_grouped}).reset_index()
df = df.merge(median_living, on = 'rooms', how = 'left') 
df['living_area'] = df['living_area'].fillna(df['median_living'])
df = df.drop(['median_living'], axis = 1)
```

<b> Я создал отдельный датафрейм, сгрупированный по количеству комнат, и рассчитанным медианным  значением жилой площади. Затем с помощью merge(), соединил его с основным датафреймом df, по столбцу 'rooms', а пустые значения в living_area заменил на медианные по количеству комнат далее дропнул из основного датафрейма добавленный столбец. 

```python
df.duplicated().sum()
# явных дубликатов в датафрейме нет 
df.head()
```

```python
df['last_price'].astype('int') 
# Целочисленное значение цены квартиры более удобно лично для меня.
df['total_images'].astype('int')
# Сложно было представить себе 2.2 фотографии. Поэтому также привел к целочисленным.
df['rooms'].astype('int')
# 3.5 комнаты, может быть и существует, но считаю подобное выражение количества комнат некорректным
df['floors_total'].astype('int')
# также как и с комнатами
df['floor'].astype('int')
# также как и с общим количеством этажей
df['balcony'].astype('int')
# также как и с комнатами. 
```

<b> Писать код диаграммы размаха для каждого столбца не очень удобно, поэтому создадим функцию, которая выведет диаграмму размаха с наименьшим значением выбросов и наиболее приближенные к усам

```python
def filtred_hist(row, column):
    q1 = row[column].quantile(0.25)
    q3 = row[column].quantile(0.75)
    iqr = q3 - q1
    result = (df[column] >= (q1 - 1.5*iqr)) & (df[column] <= (q3 + 1.5*iqr))
    return df.loc[result].boxplot([column], figsize=(15,5))
filtred_hist(df, 'total_area')
plt.show()
```

<b> Медианное значение квадратуры квартир из предоставленных данных находится в районе 50 квадратных метров. Сам ящик охватывает значения от, примерно, 40 до 70 квадратных метров. 

```python
df['total_area'].hist(bins = 30, figsize = (15,5), range = (10, 120))
plt.xlabel('Квадратура')
plt.ylabel('Количество квартир')
plt.title('Площадь квартир')
plt.show()
```

### Выберите 10 населённых пунктов с наибольшим числом объявлений. Посчитайте среднюю цену квадратного метра в этих населённых пунктах. Выделите населённые пункты с самой высокой и низкой стоимостью жилья. Эти данные можно найти по имени в столбце locality_name.

```python
df_filter['locality_name'] = df_filter['locality_name'].str.replace('посёлок ', '')
df_filter['locality_name'] = df_filter['locality_name'].str.replace('деревня ', '')
local_pivot = df_filter.pivot_table(index = 'locality_name', values = 'price_per_metr', aggfunc = ['count', 'mean'])
local_pivot.columns = ['count', 'mean']
local_pivot['mean'] = round(local_pivot['mean'], 2)
local_pivot = local_pivot.sort_values(by='count', ascending = False).head(10)
labels = ['Санкт-Петербург', 'Мурино', 'Кудрово', 'Шушары', 'Всеволожск',
          'Парголово', 'Пушкин', 'Колпино', 'Гатчина', 'Выборг']
plt.pie(local_pivot['mean'], labels = labels, autopct='%1.1f%%', rotatelabels=True, radius = 2 )
plt.show()
```

```python
high_local = local_pivot.sort_values(by='mean', ascending = False).head(3)
low_local = local_pivot.sort_values(by='mean', ascending = False).tail(3)
print(high_local, '\n', low_local)
```

<b> Из выборки по населенному пункту с наибольшим количеством объявлений, я нашел среднее арифметическое значение по цене за квадратный метр в столбце 'mean', и общее количество объявлений в столбце 'count.
    <br/> Также в последнем выводе, high_local - это населенные пункты с самой высокой ценой за квадратный метр, а low_local - с самой низкой
